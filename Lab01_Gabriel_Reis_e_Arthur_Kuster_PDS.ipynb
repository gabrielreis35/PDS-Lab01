```python
import librosa
import matplotlib.pyplot as plt
import numpy as np
from librosa.display import specshow
from librosa.core import load
import IPython.display as ipd
import os
```

```python
##importa audio
y, sr = librosa.load('/content/simpsons-sax.mp3')
```

```python
plt.stem(y)
```

```python
n = len(y)

t = np.linspace(0, n / sr, n)
plt.plot(t, y)
plt.title("Sinal no Dom√≠nio do Tempo (Tempo x Amplitude)")
plt.xlabel("Tempo (s)")
plt.ylabel("Amplitude")
plt.show()
```

```python
ipd.Audio(y, rate=sr)
```

Exerc√≠cio 2

```python
pause_duration = 2
pause_samples = int(pause_duration * sr)
split_point = len(y) // 2
y_with_pause = np.concatenate((y[:split_point], np.zeros(pause_samples), y[split_point:]))
```

```python
ipd.Audio(y_with_pause, rate=sr)
```

Exer√≠cio 3

```python
delay_seg = 0.8
decay = 0.8

delay = int(delay_seg * sr)

sinal_eco = np.zeros(len(y) + delay)
sinal_eco[:len(y)] = y
sinal_eco[delay:] += decay * y
```

```python
ipd.Audio(sinal_eco, rate=sr)
```
